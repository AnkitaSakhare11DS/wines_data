#remove outliers

```python
def determine_outlier_thresholds_iqr(dataframe, col_name, th1=0.25, th3=0.75):  #for single column
    quartile1 = dataframe[col_name].quantile(th1)
    quartile3 = dataframe[col_name].quantile(th3)
    iqr = quartile3 - quartile1
    upper_limit = quartile3 + 1.5 * iqr
    lower_limit = quartile1 - 1.5 * iqr
    return lower_limit, upper_limit
```

```python
sugar_IQR_lower,sugar_IQR_upper=determine_outlier_thresholds_iqr(Total_dataset,'residual.sugar')
print(sugar_IQR_lower,sugar_IQR_upper)
```

```python
sulfur_dioxide_IQR_lower,sulfur_dioxide_IQR_upper=determine_outlier_thresholds_iqr(Total_dataset,'free.sulfur.dioxide')
print(sulfur_dioxide_IQR_lower,sulfur_dioxide_IQR_upper)
```

```python

T_sulfur_dioxide_IQR_lower,Tsulfur_dioxide_IQR_upper=determine_outlier_thresholds_iqr(Total_dataset,'total.sulfur.dioxide')
print(T_sulfur_dioxide_IQR_lower,Tsulfur_dioxide_IQR_upper)
```

There are 2 most popular techq.
1. Label Encoder
2. One Hot encoder

1. Label Encoder :  
 Label Encoding assigns a unique number to each category

  Label encoding is suitable for categorical data where there is an inherent order or ranking among the categories.
  ex. height -> tall ,medium,short or colors

  It assigns values like
                        tall - 0
                        medium - 1
                        small - 2

          
  Income Levels:Low < Medium < High
    
  it is imported from sklearn libr.

    from sklearn.preprocessing import LabelEncoder
    encoder = LabelEncoder()
    encoded_values = encoder.fit_transform(dataframe['column'])


2. One Hot Encoding:

One-Hot Encoding is another technique to convert categorical data into a numerical form. Instead of giving each category a single number, it creates a new binary column (1 or 0) for each category.

    pd.getdummies(dataframe,columns=df['column'])
